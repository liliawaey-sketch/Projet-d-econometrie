# Pre-processing 

```python
import pandas as pd

# -----------------------------
# 1) Fichiers (dans le dossier courant)
# -----------------------------
FACTORS_CSV = "F-F_Research_Data_5_Factors_2x3.csv"
PORTF_CSV   = "6_Portfolios_ME_INV_2x3.csv"
OUT_XLSX    = "data_FF_PF.xlsx"

# -----------------------------
# 2) Charger
# -----------------------------
factors = pd.read_csv(FACTORS_CSV)
portf   = pd.read_csv(PORTF_CSV)

# -----------------------------
# 3) Garder colonnes utiles + renommer
# -----------------------------
factors = factors[["Date", "Mkt-RF", "SMB", "HML", "RMW", "CMA", "RF"]].rename(columns={"Mkt-RF": "Mkt_RF"})
portf   = portf[["Date", "SMALL LoINV", "BIG LoINV"]].rename(columns={"SMALL LoINV": "SMALL_LoINV",
                                                                      "BIG LoINV":   "BIG_LoINV"})

# -----------------------------
# 4) Parser Date (YYYYMM -> YYYY-MM-01)
# -----------------------------
def parse_yyyymm_date(s):
    s = s.astype(str).str.replace(r"\D", "", regex=True).str.slice(0, 6)  # garde YYYYMM
    return pd.to_datetime(s + "01", format="%Y%m%d", errors="coerce")

factors["Date"] = parse_yyyymm_date(factors["Date"])
portf["Date"]   = parse_yyyymm_date(portf["Date"])

# -----------------------------
# 5) Convertir en numériques (gère virgules)
# -----------------------------
def to_num(df, cols):
    for c in cols:
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(",", ".", regex=False), errors="coerce")
    return df

factors = to_num(factors, ["Mkt_RF", "SMB", "HML", "RMW", "CMA", "RF"])
portf   = to_num(portf, ["SMALL_LoINV", "BIG_LoINV"])

# -----------------------------
# 6) Nettoyer (dates valides + pas de -99.99)
# -----------------------------
factors = factors.dropna(subset=["Date"])
portf   = portf.dropna(subset=["Date"])

for df in [factors, portf]:
    df.replace([-99.99, -999], pd.NA, inplace=True)

factors = factors.dropna()
portf   = portf.dropna()

# -----------------------------
# 7) Fusion + rendements excédentaires
# -----------------------------
merged = portf.merge(factors, on="Date", how="inner")

merged["SMALL_LoINV_RF"] = merged["SMALL_LoINV"] - merged["RF"]
merged["BIG_LoINV_RF"]   = merged["BIG_LoINV"]   - merged["RF"]

# -----------------------------
# 8) Filtrer période (ex: 1990-09 à 2025-12) + rendre Date "propre"
# -----------------------------
merged["Date"] = pd.to_datetime(merged["Date"])

merged = merged[
    (merged["Date"] >= "1990-09-01") &
    (merged["Date"] <= "2025-12-01")
].copy()

merged = merged.sort_values("Date").set_index("Date")

# -----------------------------
# 9) Export
# -----------------------------
merged.reset_index().to_excel(OUT_XLSX, index=False)
```
